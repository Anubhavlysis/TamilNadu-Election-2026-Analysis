```python
#creating competitive categories

def margin_bucket(x):
    if x < 5:
        return "Very Competitive (<5%)"
    elif x < 10:
        return "Competitive (5-10%)"
    elif x < 20:
        return "Moderate (10-20%)"
    else:
        return "Safe (>20%)"
```

```python
margin_2026["competitive_type"]=(margin_2026["margin%"].apply(margin_bucket))
```

```python
margin_2021["competitive_type"]=(margin_2021["vote_margin%"].apply(margin_bucket))
```

##### - Chennai Metro exhibited significantly larger victory margins than any other region.
- Chennai Metro experienced extensive winner changes, yet the winning candidates often secured relatively comfortable victory margins.
- Outside Chennai Metro, victory margins were remarkably consistent across Tamil Nadu's regions, generally ranging between 5% and 7%.

### Insights

- Average victory margins narrowed from 11% in 2021 to 7% in 2026, indicating that constituencies were generally decided by closer contests in 2026.
- The number of constituencies where the winner secured an outright majority (>50% of votes) declined sharply from 70 to 13.
- Constituencies won with less than 35% vote share increased from 2 to 64, indicating a rise in fragmented electoral outcomes.

#### To tell this in the presentation, we can say that --

- The number of constituencies where the winner secured more than 50% of the total votes declined sharply. In 2021, there were 70 constituencies where the candidate secure more than 50% of the votes. And in 2026, that constituency decreased to 13 and it is 81.42%.
- Whereas the number of constituencies where the winner secured less than 35% of the total votes was increeased by 32 times in 2026 as compared to 2021. In 2021, there were only 2 constituency where the winning candidate had less than 35% vote share. And in 2026, there were 64 constituency were winning candidate had less than 35% of total vote share. This indicates a fragmented electoral outcome.
- Fragmented electoral outcome simply means votes are spread across many parties/candidates instead being concentrated on one or two.
- So, here we can say that 2026 elections where less fragmented as compared to 2021 elections.


- The 2026 election was characterized by narrower victory margins, fewer majority-winning constituencies, and a larger number of fragmented contests compared with 2021.
